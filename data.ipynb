```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/all_real_estate_data1.csv')
```

```python
# Вывод общей информации о датасете
print("="*50)
print("ОБЩАЯ ИНФОРМАЦИЯ О ДАТАСЕТЕ")
print("="*50)
print(f"Количество записей (объявлений): {len(df)}")
print(f"Количество признаков (колонок): {len(df.columns)}")
print("\nНазвания колонок:")
for col in df.columns:
    print(f"  - {col}")
```

```python
# Вывод информации о типах данных и пропусках
print("\n" + "="*50)
print("ТИПЫ ДАННЫХ И ПРОПУСКИ")
print("="*50)
print(df.info())
```

```python
# Описание числовых признаков
print("\n" + "="*50)
print("СТАТИСТИКА ПО ЧИСЛОВЫМ ПРИЗНАКАМ")
print("="*50)
numeric_cols = ['price_numeric', 'total_area_numeric', 'price_per_meter_numeric']
print(df[numeric_cols].describe())
```

```python
# Анализ распределения цен
print("\n" + "="*50)
print("РАСПРЕДЕЛЕНИЕ ЦЕН НА НЕДВИЖИМОСТЬ")
print("="*50)
print(f"Минимальная цена: {df['price_numeric'].min():,.0f} ₽")
print(f"Максимальная цена: {df['price_numeric'].max():,.0f} ₽")
print(f"Средняя цена: {df['price_numeric'].mean():,.0f} ₽")
print(f"Медианная цена: {df['price_numeric'].median():,.0f} ₽")
```

```python
# Анализ самых частотных категорий
print("\n" + "="*50)
print("ТОП-5 САМЫХ ЧАСТОТНЫХ ЗНАЧЕНИЙ В КАТЕГОРИАЛЬНЫХ ПРИЗНАКАХ")
print("="*50)
categorical_cols = ['property_type', 'building_type', 'heating', 'renovation', 'deal_terms']
for col in categorical_cols:
    if col in df.columns:
        print(f"\n{col.upper()}:")
        print(df[col].value_counts().head(5).to_string())
```

```python
# Анализ пропусков в ключевых полях
print("\n" + "="*50)
print("ПРОЦЕНТ ПРОПУСКОВ В КЛЮЧЕВЫХ ПОЛЯХ")
print("="*50)
for col in ['living_area', 'kitchen_area', 'floor', 'year_built', 'mortgage']:
    if col in df.columns:
        missing_pct = (df[col].isna().sum() / len(df)) * 100
        print(f"{col}: {missing_pct:.1f}% пропусков")
```

```python
# Вывод первых 5 строк для примера
print("\n" + "="*50)
print("ПРИМЕР ПЕРВЫХ 5 ОБЪЯВЛЕНИЙ")
print("="*50)
print(df[['title', 'price', 'total_area', 'address']].head().to_string())
```
